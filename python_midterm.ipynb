```python
import requests
import json
import re
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from time import sleep
from tqdm import tqdm
```

```python
class DiscWorld:
    def __init__(self):
        self.link = 'http://wiki.lspace.org/mediawiki/List_of_Pratchett_characters'
    
    def getSoup(self, link):
        #borrowed this from J. Sila
        r = requests.get(link)
        r.encoding = 'UTF-8'
        return BeautifulSoup(r.text,'lxml')
    
    def characters(self):
        soup = self.getSoup(self.link)
        find = soup.find('span',{'id':'Unseen_University_and_the_Wizards'}).parent.previous.previous.parent.parent.parent.find_all('li')
        return find
    
    def listLinks(self):
        list_people = []
        links = self.characters()
        for person in links:
            person_link = person.find('a', href=True)['href']
            list_people.append('http://wiki.lspace.org' + person_link)
        return list_people
    
    def peopleBooks(self):
        ppl_book_dict = {}
        for link in self.listLinks():
            soup = self.getSoup(link)
            name = soup.find('h1', {'id':'firstHeading'}).text
            books = soup.find_all('a', title = lambda x: x and 'Book:' in x)
            list_books = []
            for element in books:
                title = re.sub('Book:', '', element['title'])
                list_books.append(title)
            unique_list = list(set(list_books))
            unique_books = ', '.join(unique_list)
            ppl_book_dict[name] = unique_books
        return ppl_book_dict
    
    def makeDf(self):
        df = pd.DataFrame.from_dict(self.peopleBooks(), orient='index')
        df['Names'] = df.index
        #CANNOT RENAME - I do not why this does not work:
        df = df.rename(columns={'0': 'Books'})
        return df
                
            
        
```

```python
inst = DiscWorld()
df = inst.makeDf()
df
```
